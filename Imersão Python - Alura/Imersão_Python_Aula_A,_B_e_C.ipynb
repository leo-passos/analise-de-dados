```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel(r'/content/Imersão Python - Tabela de ações.xlsx',sheet_name='Principal')

df_principal.head(5)
```

```python
df_total_acoes = pd.read_excel(r'/content/Imersão Python - Tabela de ações.xlsx',sheet_name='Total_de_acoes')

df_total_acoes.head(5)
```

```python
df_ticker = pd.read_excel(r'/content/Imersão Python - Tabela de ações.xlsx', sheet_name="Ticker")
df_ticker.head(5)
```

```python
df_chatgpt = pd.read_excel(r'/content/Imersão Python - Tabela de ações.xlsx', sheet_name="ChatGPT")
df_chatgpt.head(5)
```

```python
# prompt: Usando o DataFrame df_chatgpt: Quero pegra os nomes dos colunas do dataframe df_principal

list(df_principal.columns)
```

```python
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()

df_principal.head(5)
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()

df_principal.head(5)
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final']/(df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal.head(5)
```

```python
df_principal = df_principal.drop(columns={'Código'})
df_principal.head(5)
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['Qtde. Teórica']
df_principal.head(5)
```

```python
pd.options.display.float_format = '{:.2f}'.format

df_principal.head(5)
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal.head(5)
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtde_teorica'}).copy()

df_principal.head(5)
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))

df_principal

```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')

df_principal = df_principal.drop(columns={'Ticker'})

df_principal.head(5)
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Empresas', how='left')
df_principal = df_principal.rename(columns={'Seguimento':'Segmento'}).copy()
df_principal = df_principal.drop(columns={'Empresas'})

df_principal.head(5)
```

```python
df_principal = df_principal.rename(columns={'Idade (aprox.)':'Ano_de_fundacao'}).copy()

df_principal.head(5)
```

```python
df_principal ['Cat_idade'] = df_principal['Ano_de_fundacao'].apply(lambda x: 'Mais de 100 anos' if x > 100 else ('Menos de 50 anos' if x < 50 else 'Entre 50 e 100 anos'))

df_principal
```

```python
maior = df_principal['Variacao_rs'].max()
menor = df_principal['Variacao_rs'].min()
media = df_principal['Variacao_rs'].mean()
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

df_resultado = pd.DataFrame({
    'Maior': [maior],
    'Menor': [menor],
    'Média': [media],
    'Média de quem subiu': [media_subiu],
    'Média de quem desceu': [media_desceu]
})

df_resultado = df_resultado.transpose()

df_resultado.columns = ['Resultados']

df_resultado
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']

df_principal_subiu.head(5)
```

```python
df_analise_seguimento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_seguimento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x= 'Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação (R$ por resultados)')

fig.show()
```
